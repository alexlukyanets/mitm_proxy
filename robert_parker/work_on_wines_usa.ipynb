```python
import pandas as pd
from database.mysql_connection_string import mysql_connection_string
import sqlalchemy as sa
from sqlalchemy.sql import select
from database.models.parker_wine import ParkerWine
from sqlalchemy.dialects import mysql
```

```python
engine = sa.create_engine(mysql_connection_string()).connect()
```

```python
stmt = select(ParkerWine)
stmt_compiled = stmt.compile(compile_kwargs={"literal_binds": True}, dialect=mysql.dialect())
results = engine.execute(str(stmt_compiled))
```

```python
wines = pd.DataFrame(results.fetchall())
wines
```

```python
wines = wines.drop(['id', 'created_at', 'updated_at'], axis=1)
```

```python
wines = wines.drop_duplicates()
```

```python
usa_wines = wines[wines['country'] == "USA"]
usa_wines
```

```python
usa_wines = usa_wines.rename(columns={"wine_id": "vintage_id"})
usa_wines
```

```python
usa_wines.source_link[0]
```

```python
usa_wines['source_link'] = usa_wines.apply(lambda row: "https://www.robertparker.com"+ row.source_link, axis=1)
usa_wines['reviewer_url'] = usa_wines.apply(lambda row: "https://www.robertparker.com/about/reviewers/"+ row.reviewer_id, axis=1)
usa_wines['vintage_url'] = usa_wines.apply(lambda row: "https://www.robertparker.com/about/reviewers/"+ row.vintage_id, axis=1)
```

```python
usa_wines.columns
```

```python
columns_order = ['vintage_url', 'reviewer_url', 'source_link','country','locale', 'location_name', 'region', 'site',
                 'vintage_id', 
                  'maturity',  'type', 'color',
       'variety', 'sweetness', 'rating',  'price', 'vintage',
       'testing_note_id', 'name', 'name_display', 'producer_name',
       'producer_show',  'source_text',
       'reviewer', 'reviewer_id', 'drink_date', 'issue_date',
       'producer_id',  'description']
```
